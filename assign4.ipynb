```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("loan_data.csv")
df.head()
```

```python
from sklearn.datasets import make_classification

X, y = make_classification(
    n_features=6,
    n_classes=3,
    n_samples=800,
    n_informative=2,
    random_state=1,
    n_clusters_per_class=1,
)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1], c=y, marker="*");
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=df,x='purpose',hue='not.fully.paid')
plt.xticks(rotation=45, ha='right');
```

```python
pre_df = pd.get_dummies(df,columns=['purpose'],drop_first=True)
pre_df.head()
```

```python
from sklearn.model_selection import train_test_split

X = pre_df.drop('not.fully.paid', axis=1)
y = pre_df['not.fully.paid']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=125
)
```

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()

model.fit(X_train, y_train);
```

```python
from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    ConfusionMatrixDisplay,
    f1_score,
    classification_report,
)

y_pred = model.predict(X_test)

accuray = accuracy_score(y_pred, y_test)
f1 = f1_score(y_pred, y_test, average="weighted")

print("Accuracy:", accuray)
print("F1 Score:", f1)
```

```python
labels = ["Fully Paid", "Not fully Paid"]
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
disp.plot();
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate some data
data = np.random.normal(loc=0, scale=1, size=1000)  # Example data, replace this with your actual data

# Fit a normal distribution to the data
mu, std = norm.fit(data)

# Plot the histogram
plt.figure(figsize=(8, 6))
sns.histplot(data, kde=True, bins=30, color='blue', label='Data histogram')

# Plot the PDF (Probability Density Function)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2, label='Normal distribution curve')

plt.title('Normal Distribution Curve')
plt.xlabel('Data')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate some data
data_mean = 0  # Mean of the distribution
data_std = 1   # Standard deviation of the distribution
data_range = np.linspace(data_mean - 3*data_std, data_mean + 3*data_std, 1000)  # Range of values

# Calculate the probability density function (PDF) for the given range
pdf_values = norm.pdf(data_range, data_mean, data_std)

# Plot the Gaussian distribution curve
plt.figure(figsize=(8, 6))
plt.plot(data_range, pdf_values, color='blue', lw=2, label='Gaussian Distribution')

# Add labels and title
plt.title('Gaussian Distribution Curve')
plt.xlabel('Values')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```
